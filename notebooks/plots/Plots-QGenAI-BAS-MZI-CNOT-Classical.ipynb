# Combined QGenAI BAS plots

Notebook này đọc kết quả đã lưu từ:

- `QGENAI-BAS-MERLIN-MZI-1layers-history.json` ... `QGENAI-BAS-MERLIN-MZI-3layers-history.json`
- `QGENAI-BAS-MERLIN-CNOT-1layers-history.json` ... `QGENAI-BAS-MERLIN-CNOT-3layers-history.json`
- `QGENAI-BAS-MERLIN-CLASSICAL-1layers-history.json` ... `QGENAI-BAS-MERLIN-CLASSICAL-3layers-history.json`

Mục tiêu là gom các plot rời trong notebook training thành các dashboard chung để so sánh MZI, repeated CNOT, và Classical generator trên BAS 2x2.


```python

import json
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

try:
    import pandas as pd
except ImportError:
    pd = None

try:
    from IPython.display import display
except ImportError:
    def display(value):
        print(value)


# ============================================================
# Configuration
# ============================================================

HISTORY_DIR = Path(r"C:\Quantum")
OUTPUT_DIR = HISTORY_DIR

LAYERS = [1, 2, 3]
BIT_WIDTH = 4
GRID_SHAPE = (2, 2)

MODEL_SPECS = [
    {
        "key": "MZI",
        "label": "MZI",
        "color": "#0072B2",
        "filename": lambda layer: f"QGENAI-BAS-MERLIN-MZI-{layer}layers-history.json",
    },
    {
        "key": "CNOT",
        "label": "Repeated CNOT",
        "color": "#D55E00",
        "filename": lambda layer: f"QGENAI-BAS-MERLIN-CNOT-{layer}layers-history.json",
    },
    {
        "key": "CLASSICAL",
        "label": "Classical MLP",
        "color": "#009E73",
        "filename": lambda layer: f"QGENAI-BAS-MERLIN-CLASSICAL-{layer}layers-history.json",
    },
]

MODEL_BY_KEY = {spec["key"]: spec for spec in MODEL_SPECS}
STATE_LABELS = [format(index, f"0{BIT_WIDTH}b") for index in range(2 ** BIT_WIDTH)]

EXPORT_DPI = 600
LEGEND_FONTSIZE = 16
plt.rcParams.update({
    "font.size": 15,
    "axes.titlesize": 20,
    "axes.labelsize": 17,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
    "legend.fontsize": LEGEND_FONTSIZE,
    "legend.title_fontsize": LEGEND_FONTSIZE,
    "figure.titlesize": 30,
    "lines.linewidth": 2.6,
})


def safe_float(value):
    if value is None:
        return np.nan
    try:
        return float(value)
    except (TypeError, ValueError):
        return np.nan


def maybe_savefig(fig, filename):
    path = OUTPUT_DIR / filename
    fig.savefig(path, dpi=EXPORT_DPI, bbox_inches="tight")
    pdf_path = path.with_suffix(".pdf")
    fig.savefig(pdf_path, bbox_inches="tight")
    print("Saved figure:", path)
    print("Saved vector figure:", pdf_path)


def state_to_image(state):
    bits = [int(bit) for bit in format(int(state), f"0{BIT_WIDTH}b")]
    return np.array(bits, dtype=float).reshape(GRID_SHAPE)


def load_histories():
    histories = {}
    missing = []

    for spec in MODEL_SPECS:
        for layer in LAYERS:
            path = HISTORY_DIR / spec["filename"](layer)
            if not path.exists():
                missing.append(str(path))
                continue
            with open(path, "r", encoding="utf-8") as file:
                history = json.load(file)
            history["_model_key"] = spec["key"]
            history["_model_label"] = spec["label"]
            history["_color"] = spec["color"]
            history["_path"] = str(path)
            history["_layer"] = int(history.get("layers", layer))
            histories[(spec["key"], layer)] = history

    print(f"Loaded histories: {len(histories)} / {len(MODEL_SPECS) * len(LAYERS)}")
    if missing:
        print("Missing files:")
        for path in missing:
            print(" -", path)
    return histories


histories = load_histories()


def get_target_distribution():
    for history in histories.values():
        target = history.get("final", {}).get("target_distribution")
        if target is not None:
            return np.array(target, dtype=float)

    # Fallback BAS 2x2 target: all rows or all columns are constant.
    target = np.zeros(2 ** BIT_WIDTH, dtype=float)
    bas = []
    for index in range(2 ** BIT_WIDTH):
        image = state_to_image(index)
        row_constant = np.all(image == image[:, :1], axis=1).all()
        col_constant = np.all(image == image[:1, :], axis=0).all()
        if row_constant or col_constant:
            bas.append(index)
    target[bas] = 1.0 / len(bas)
    return target


TARGET_DISTRIBUTION = get_target_distribution()
TARGET_SUPPORT = TARGET_DISTRIBUTION > 0
TARGET_SUPPORT_INDICES = np.flatnonzero(TARGET_SUPPORT)


# ============================================================
# Tables
# ============================================================

def final_metric(history, metric):
    if history is None:
        return np.nan
    if metric == "training_time_s":
        return safe_float(history.get("timing", {}).get("training_time_s"))
    if metric == "trainable_parameters":
        return safe_float(history.get("trainable_parameters"))
    return safe_float(history.get("final", {}).get(metric))


def build_summary_rows(histories):
    rows = []
    for spec in MODEL_SPECS:
        for layer in LAYERS:
            history = histories.get((spec["key"], layer))
            if history is None:
                continue
            final = history.get("final", {})
            rows.append({
                "model": spec["label"],
                "model_key": spec["key"],
                "layer": layer,
                "loss": safe_float(final.get("loss")),
                "js_divergence": safe_float(final.get("js_divergence")),
                "total_variation_distance": safe_float(final.get("total_variation_distance")),
                "negative_log_likelihood": safe_float(final.get("negative_log_likelihood")),
                "target_support_probability": safe_float(final.get("target_support_probability")),
                "mode_coverage": safe_float(final.get("mode_coverage")),
                "target_support_modes": safe_float(final.get("target_support_modes")),
                "entropy": safe_float(final.get("entropy")),
                "logical_success_mass": safe_float(final.get("logical_success_mass")),
                "raw_result_count": safe_float(final.get("raw_result_count")),
                "training_time_s": safe_float(history.get("timing", {}).get("training_time_s")),
                "mean_step_time_s": safe_float(history.get("timing", {}).get("mean_step_time_s")),
                "mean_epoch_time_s": safe_float(history.get("timing", {}).get("mean_epoch_time_s")),
                "trainable_parameters": safe_float(history.get("trainable_parameters")),
                "learning_rate": safe_float(history.get("learning_rate")),
                "optimizer": history.get("optimizer"),
                "simulator": history.get("simulator", "classical"),
                "encoding": history.get("encoding", "none"),
                "history_path": history.get("_path", ""),
            })
    return rows


def build_curve_rows(histories):
    rows = []
    for spec in MODEL_SPECS:
        for layer in LAYERS:
            history = histories.get((spec["key"], layer))
            if history is None:
                continue
            x_key = "step" if "step" in history else "epoch"
            x_values = history.get(x_key, [])
            for i, x_value in enumerate(x_values):
                row = {
                    "model": spec["label"],
                    "model_key": spec["key"],
                    "layer": layer,
                    "iteration_type": x_key,
                    "iteration": int(x_value),
                }
                for metric in [
                    "loss",
                    "js_divergence",
                    "total_variation_distance",
                    "negative_log_likelihood",
                    "target_support_probability",
                    "mode_coverage",
                    "entropy",
                    "grad_norm",
                    "step_time_s",
                    "epoch_time_s",
                ]:
                    series = history.get(metric, [])
                    row[metric] = safe_float(series[i]) if i < len(series) else np.nan
                rows.append(row)
    return rows


def build_distribution_rows(histories):
    rows = []
    for spec in MODEL_SPECS:
        for layer in LAYERS:
            history = histories.get((spec["key"], layer))
            if history is None:
                continue
            generated = np.array(history.get("final", {}).get("generated_distribution", []), dtype=float)
            target = np.array(history.get("final", {}).get("target_distribution", TARGET_DISTRIBUTION), dtype=float)
            n = min(len(generated), len(target), len(STATE_LABELS))
            for state in range(n):
                rows.append({
                    "model": spec["label"],
                    "model_key": spec["key"],
                    "layer": layer,
                    "state": state,
                    "bitstring": STATE_LABELS[state],
                    "target_probability": float(target[state]),
                    "generated_probability": float(generated[state]),
                    "probability_error": float(generated[state] - target[state]),
                    "absolute_error": float(abs(generated[state] - target[state])),
                    "is_target_pattern": bool(target[state] > 0),
                })
    return rows


summary_rows = build_summary_rows(histories)
curve_rows = build_curve_rows(histories)
distribution_rows = build_distribution_rows(histories)

if pd is not None:
    summary_df = pd.DataFrame(summary_rows)
    curve_df = pd.DataFrame(curve_rows)
    distribution_df = pd.DataFrame(distribution_rows)

    summary_path = OUTPUT_DIR / "QGENAI-BAS-MERLIN-COMBINED-summary.csv"
    curve_path = OUTPUT_DIR / "QGENAI-BAS-MERLIN-COMBINED-curves.csv"
    distribution_path = OUTPUT_DIR / "QGENAI-BAS-MERLIN-COMBINED-distributions.csv"

    summary_df.to_csv(summary_path, index=False)
    curve_df.to_csv(curve_path, index=False)
    distribution_df.to_csv(distribution_path, index=False)

    print("Saved:", summary_path)
    print("Saved:", curve_path)
    print("Saved:", distribution_path)
    display(summary_df.sort_values(["layer", "model_key"]))
else:
    print("pandas is not installed. Tables are available as Python lists.")


# ============================================================
# Plot 1: BAS target patterns
# ============================================================

def plot_target_bas_patterns():
    target_states = TARGET_SUPPORT_INDICES.tolist()
    cols = len(target_states)
    fig, axes = plt.subplots(1, cols, figsize=(1.8 * cols, 2.4))
    if cols == 1:
        axes = [axes]

    for ax, state in zip(axes, target_states):
        ax.imshow(state_to_image(state), cmap="gray_r", vmin=0, vmax=1)
        ax.set_title(STATE_LABELS[state], fontsize=13)
        ax.set_xticks([])
        ax.set_yticks([])
        for spine in ax.spines.values():
            spine.set_color("#444444")
            spine.set_linewidth(0.8)

    fig.suptitle("BAS 2x2 target support", fontsize=22)
    fig.tight_layout()
    maybe_savefig(fig, "QGENAI-BAS-target-patterns.png")
    plt.show()


# ============================================================
# Plot 2: final metrics
# ============================================================

def plot_final_metric_overview():
    metric_specs = [
        ("loss", "Objective loss", "lower is better"),
        ("js_divergence", "JS divergence", "lower is better"),
        ("total_variation_distance", "TVD", "lower is better"),
        ("target_support_probability", "Target mass", "higher is better"),
        ("mode_coverage", "Mode coverage", "higher is better"),
        ("training_time_s", "Training time (s)", "lower is better"),
    ]

    fig, axes = plt.subplots(2, 3, figsize=(24, 13))
    axes = axes.ravel()
    x = np.arange(len(LAYERS))
    bar_width = 0.24

    for ax, (metric, title, subtitle) in zip(axes, metric_specs):
        for model_index, spec in enumerate(MODEL_SPECS):
            values = [final_metric(histories.get((spec["key"], layer)), metric) for layer in LAYERS]
            offsets = x + (model_index - 1) * bar_width
            bars = ax.bar(
                offsets,
                values,
                width=bar_width,
                label=spec["label"],
                color=spec["color"],
                alpha=0.88,
            )
            for bar, value in zip(bars, values):
                if np.isnan(value):
                    continue
                text = f"{value:.2f}" if metric in ["training_time_s", "mode_coverage"] else f"{value:.3f}"
                ax.text(
                    bar.get_x() + bar.get_width() / 2,
                    bar.get_height(),
                    text,
                    ha="center",
                    va="bottom",
                    fontsize=12,
                )
        ax.set_title(f"{title}\n{subtitle}", fontsize=15)
        ax.set_xticks(x, labels=[f"{layer}L" for layer in LAYERS])
        ax.grid(axis="y", alpha=0.25)

    axes[0].legend(ncol=3, fontsize=LEGEND_FONTSIZE, frameon=True, borderpad=0.8, handlelength=2.8)
    fig.suptitle("QGenAI BAS final metrics | MZI vs repeated CNOT vs Classical", fontsize=30)
    fig.tight_layout()
    maybe_savefig(fig, "QGENAI-BAS-COMBINED-final-metrics.png")
    plt.show()


# ============================================================
# Plot 3: training curves
# ============================================================

def plot_loss_convergence():
    fig, axes = plt.subplots(1, len(LAYERS), figsize=(7.2 * len(LAYERS), 6.2), sharex=False, sharey=True)
    if len(LAYERS) == 1:
        axes = [axes]

    for col, layer in enumerate(LAYERS):
        ax = axes[col]
        for spec in MODEL_SPECS:
            history = histories.get((spec["key"], layer))
            if history is None:
                continue

            x_key = "step" if "step" in history else "epoch"
            x = np.array(history.get(x_key, []), dtype=float)
            y = np.array(history.get("loss", []), dtype=float)
            if len(x) == 0 or len(y) == 0:
                continue

            n = min(len(x), len(y))
            linestyle = "--" if spec["key"] == "CLASSICAL" else "-"
            ax.plot(
                x[:n],
                y[:n],
                color=spec["color"],
                linestyle=linestyle,
                linewidth=3.0,
                alpha=0.9,
                label=spec["label"],
            )

        ax.set_title(f"{layer} layer(s)", fontsize=21)
        ax.set_xlabel("Step / epoch")
        ax.set_ylabel("Objective loss")
        ax.tick_params(axis="y", labelleft=True)
        ax.tick_params(axis="both", labelsize=12)
        ax.grid(alpha=0.25)
        if col == 0:
            ax.legend(fontsize=LEGEND_FONTSIZE, frameon=True, borderpad=0.8, handlelength=2.8)

    fig.suptitle("Loss convergence | QGenAI BAS | 1-3 layers | all models", fontsize=32)
    fig.tight_layout()
    maybe_savefig(fig, "QGENAI-BAS-COMBINED-loss-convergence.png")
    plt.show()


# ============================================================
# Plot 4: target vs generated distributions
# ============================================================

def plot_combined_distribution_overlays():
    fig, axes = plt.subplots(len(LAYERS), len(MODEL_SPECS), figsize=(30, 17), sharey=True, squeeze=False)
    states = np.arange(len(STATE_LABELS))
    global_ymax = 0.0
    for layer in LAYERS:
        for spec in MODEL_SPECS:
            history = histories.get((spec["key"], layer))
            if history is None:
                continue
            final = history.get("final", {})
            generated = np.array(final.get("generated_distribution", []), dtype=float)
            target = np.array(final.get("target_distribution", TARGET_DISTRIBUTION), dtype=float)
            if len(generated) or len(target):
                global_ymax = max(global_ymax, float(np.max(generated)) if len(generated) else 0.0)
                global_ymax = max(global_ymax, float(np.max(target)) if len(target) else 0.0)
    global_ymax = max(0.20, global_ymax * 1.18)

    for row, layer in enumerate(LAYERS):
        for col, spec in enumerate(MODEL_SPECS):
            ax = axes[row, col]
            history = histories.get((spec["key"], layer))
            if history is None:
                ax.axis("off")
                continue

            final = history.get("final", {})
            generated = np.array(final.get("generated_distribution", []), dtype=float)
            target = np.array(final.get("target_distribution", TARGET_DISTRIBUTION), dtype=float)
            n = min(len(generated), len(target), len(states))

            ax.bar(states[:n] - 0.18, target[:n], width=0.36, color="black", alpha=0.28, label="Target")
            ax.bar(states[:n] + 0.18, generated[:n], width=0.36, color=spec["color"], alpha=0.86, label="Generated")
            ax.set_title(
                f"{spec['label']} | {layer}L\n"
                f"loss={final.get('loss', np.nan):.3f}, TVD={final.get('total_variation_distance', np.nan):.3f}",
                fontsize=18,
            )
            ax.set_xticks(states[:n], labels=STATE_LABELS[:n], rotation=90, fontsize=11)
            ax.set_ylim(0.0, global_ymax)
            ax.set_ylabel("Probability")
            ax.tick_params(axis="y", labelleft=True)
            ax.tick_params(axis="both", labelsize=12)
            ax.grid(axis="y", alpha=0.20)
            if row == 0 and col == 0:
                ax.legend(fontsize=18, frameon=True, borderpad=0.8, handlelength=2.8)

    fig.suptitle("Target vs generated BAS distribution | 3 layers x 3 models", fontsize=32)
    fig.tight_layout()
    maybe_savefig(fig, "QGENAI-BAS-COMBINED-target-vs-generated.png")
    plt.show()


# ============================================================
# Plot 5: top generated BAS-like images
# ============================================================

def plot_top_generated_patterns_grid(top_k=6):
    rows = len(LAYERS) * len(MODEL_SPECS)
    fig, axes = plt.subplots(rows, top_k, figsize=(1.85 * top_k, 1.55 * rows), squeeze=False)

    row_index = 0
    for layer in LAYERS:
        for spec in MODEL_SPECS:
            history = histories.get((spec["key"], layer))
            if history is None:
                for col in range(top_k):
                    axes[row_index, col].axis("off")
                row_index += 1
                continue

            top = history.get("final", {}).get("top_generated_patterns", [])[:top_k]
            for col in range(top_k):
                ax = axes[row_index, col]
                ax.set_xticks([])
                ax.set_yticks([])
                if col >= len(top):
                    ax.axis("off")
                    continue
                entry = top[col]
                state = int(entry.get("state", entry.get("index", int(entry.get("bitstring", "0"), 2))))
                prob = float(entry["generated_probability"])
                mark = "target" if entry.get("is_target_pattern") else "off-target"
                ax.imshow(state_to_image(state), cmap="gray_r", vmin=0, vmax=1)
                ax.set_title(f"{STATE_LABELS[state]}\n{prob:.3f}\n{mark}", fontsize=12)
                for spine in ax.spines.values():
                    spine.set_color(spec["color"])
                    spine.set_linewidth(1.0)
            axes[row_index, 0].set_ylabel(f"{spec['key']} {layer}L", rotation=0, labelpad=34, va="center")
            row_index += 1

    fig.suptitle(f"Top {top_k} generated patterns per model/layer", fontsize=30)
    fig.tight_layout()
    maybe_savefig(fig, "QGENAI-BAS-COMBINED-top-generated-patterns.png")
    plt.show()


# ============================================================
# Plot 6: absolute distribution error
# ============================================================

def plot_distribution_error_heatmap():
    fig, axes = plt.subplots(
        len(LAYERS),
        len(MODEL_SPECS),
        figsize=(24, 12),
        squeeze=False,
        sharex=True,
        constrained_layout=True,
    )

    max_error = 0.0
    errors = {}
    for layer in LAYERS:
        for spec in MODEL_SPECS:
            history = histories.get((spec["key"], layer))
            if history is None:
                continue
            final = history.get("final", {})
            generated = np.array(final.get("generated_distribution", []), dtype=float)
            target = np.array(final.get("target_distribution", TARGET_DISTRIBUTION), dtype=float)
            n = min(len(generated), len(target), len(STATE_LABELS))
            error = np.abs(generated[:n] - target[:n]).reshape(1, -1)
            errors[(spec["key"], layer)] = error
            max_error = max(max_error, float(np.max(error)))

    image = None
    for row, layer in enumerate(LAYERS):
        for col, spec in enumerate(MODEL_SPECS):
            ax = axes[row, col]
            error = errors.get((spec["key"], layer))
            if error is None:
                ax.axis("off")
                continue
            image = ax.imshow(error, cmap="magma", aspect="auto", vmin=0, vmax=max_error)
            ax.set_title(f"{spec['label']} | {layer}L", fontsize=15)
            ax.set_yticks([])
            ax.set_xticks(np.arange(len(STATE_LABELS)), labels=STATE_LABELS, rotation=90, fontsize=7)

    fig.suptitle("Absolute distribution error by bitstring, shared color scale", fontsize=30)
    maybe_savefig(fig, "QGENAI-BAS-COMBINED-distribution-error.png")
    plt.show()


plot_target_bas_patterns()
plot_final_metric_overview()
plot_loss_convergence()
plot_combined_distribution_overlays()
plot_top_generated_patterns_grid(top_k=6)
plot_distribution_error_heatmap()


# ============================================================
# Useful previews
# ============================================================

if pd is not None:
    display(
        distribution_df
        .sort_values(["layer", "model_key", "absolute_error"], ascending=[True, True, False])
        .groupby(["layer", "model_key"])
        .head(5)
        .reset_index(drop=True)
    )
else:
    distribution_rows[:10]
```
